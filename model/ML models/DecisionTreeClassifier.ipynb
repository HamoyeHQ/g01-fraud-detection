### FRAUD-DETECTION MACHINE LEARNING MODEL USING DECISION TREE CLASSIFIER

```python
import pandas as pd
import numpy as np
```

```python
#data = pd.read_csv('../input/creditcardfraud/creditcard.csv')
data= pd.read_csv('creditcard.csv')
data.head()
```

```python
# Removing irrelevant features

data= data.drop(columns= ["Time"])
```

```python
data.shape
```

```python
# checking the class distributions of the categorical target variable

data['Class'].value_counts()
```

```python
# sorting out instances in class 1

class_1= data[data['Class']== 1]
class_1
```

```python
# sorting out instances in class 0

class_0= data[data['Class']== 0].sample(1000) # randomly picking 1000 samples from "Class"
class_0
```

```python
# combining both classes by appending

data_df= class_1.append(class_0)
data_df
```

```python
# shuffling data_df

import sklearn.utils

data_df= sklearn.utils.shuffle(data_df)
data_df=  data_df.reset_index(drop= True)
```

```python
# checking the new shape of df

data_df.shape
```

```python
# checking the class distributions of the target

data_df['Class'].value_counts()
```

```python
# separating data_df into features and target variables

X= data_df.drop(columns= 'Class')
y= data_df['Class']
```

#### Splitting dataset into trains and tests

```python
# splitting the data into training and testing sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 0.25, random_state= 0)
```

```python
# checking the class distribution of the target variable (y_train)

y_train.value_counts()
```

#### Applying SMOTE to make classes balanced

```python
# due to the imblance, SMOTE will be used increase the number of instances in the class with the lower instances cases.
# SMOTE takes the entire dataset as an input, but it increases the percentage of only the minorty class

import imblearn
from imblearn.over_sampling import SMOTE
smote= SMOTE(random_state= 1)
X_train_balanced, y_balanced= smote.fit_sample(X_train, y_train)
```

```python
X_train_balanced
```

```python
y_balanced
```

```python
# checking the class distribution of the target classes after applying SMOTE

y_balanced.value_counts() # classes have become equal
```

#### Scaling and converting X_train_balanced to dataframe

```python
#  normalising the new balanced dataset using StandardScaler

from sklearn.preprocessing import StandardScaler
scalar= StandardScaler()
normalised_train_df= scalar.fit_transform(X_train_balanced)
normalised_train_df
```

```python
# converting to a dataframe


normalised_train_df= pd.DataFrame(normalised_train_df, columns= X_train_balanced.columns)
normalised_train_df
```

#### Scaling and converting X_test to dataframe

```python
X_test
```

```python
X_test= X_test.reset_index(drop= True)
```

```python
X_test.head(5)
```

```python
# normalising X_test using MinMaxScaler() function

normalised_test_df= scalar.transform(X_test)
```

```python
normalised_test_df
```

```python
# converting to a dataframe

normalised_test_df= pd.DataFrame(normalised_test_df, columns= X_test.columns)
normalised_test_df
```

### Modelling using Decision Tree Classifier

#### Decision Tree 

```python
from sklearn import tree
dec_tree= tree.DecisionTreeClassifier()
dec_tree.fit(normalised_train_df, y_balanced)
```

#### Confusion Matrix for Decision Tree

```python
from sklearn.metrics import confusion_matrix

new_pred= dec_tree.predict(normalised_test_df)

cnf_mat_dt= confusion_matrix(y_true= y_test, y_pred= new_pred, labels= [1, 0])

cnf_mat_dt
```

#### Accuracy

```python
from sklearn.metrics import accuracy_score

accuracy_dt= accuracy_score(y_true=y_test, y_pred= new_pred)
print('Accuracy: {}'.format(round(accuracy_dt* 100), 2))
```

#### Precision

```python
from sklearn.metrics import precision_score

precision_dt= precision_score(y_true=y_test, y_pred= new_pred, pos_label= 1)
print('Precision: {}'.format(round(precision_dt* 100), 2))
```

#### Recall

```python
from sklearn.metrics import recall_score

recall_dt= recall_score(y_true=y_test, y_pred= new_pred, pos_label= 1)
print('recall: {}'.format(round(recall_dt* 100), 2))
```

#### F1_score

```python
from sklearn.metrics import f1_score

f1_dt= f1_score(y_true= y_test, y_pred= new_pred, pos_label= 1)
print('f1_score: {}'.format(round(f1_dt* 100), 2))
```

### Performing Cross Validation

#### 1). K-Fold Cross Validation

```python
from sklearn.model_selection import KFold
from sklearn.metrics import f1_score

kf= KFold(n_splits= 4)
kf.split(normalised_train_df)
f1_scores1= []

# run for every split

for train_index, test_index in kf.split(normalised_train_df):
    X_train, X_test= normalised_train_df.iloc[train_index], normalised_train_df.iloc[test_index]
    
    y_train, y_test= y_balanced[train_index], y_balanced[test_index]
    
    model= dec_tree.fit(X_train, y_train)
    
    # save result to list
    
    f1_scores1.append(f1_score(y_true= y_test, y_pred=model.predict(X_test), pos_label= 1)*100)
    
print("The f1 scores for k-fold cv are ", f1_scores1)

#Average k-fold f1 score
av1 = np.mean(f1_scores1)
print("The average f1 score for k-fold cv is ", av1)
    
```

#### 2).StratifiedKFold cross validation

```python
from sklearn.model_selection import StratifiedKFold
import numpy as np

skf= StratifiedKFold(n_splits= 4, shuffle= True, random_state= 1)
f1_scores2= []

#run for every split

for train_index, test_index in skf.split(normalised_train_df, y_balanced):
    X_train, X_test= np.array(normalised_train_df)[train_index], np.array(normalised_train_df)[test_index]
    
    y_train, y_test= y_balanced[train_index], y_balanced[test_index]
    
    model= dec_tree.fit(X_train, y_train)
    
    #save result to list
    
    f1_scores2.append(f1_score(y_true= y_test, y_pred=model.predict(X_test), pos_label= 1))


print("The f1 scores for stratified k-fold cv are ", f1_scores2)

#Average stratified k-fold f1 score
av2 = np.mean(f1_scores2)
print("The average f1 score for stratified k-fold cv is ", av2)
```

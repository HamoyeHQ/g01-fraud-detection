```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import IsolationForest

from sklearn.metrics import f1_score

df=pd.read_csv('creditcard.csv')
```

```python
df
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(df.corr(), annot = True)

plt.show()
```

```python
# V1 to V28 are the most important features according to the PCA done on  this data. Dropping others.
# df.drop(columns=['Amount','Time'], axis=1, inplace=True)
```

```python
X=df.drop(columns=['Class','Amount','Time'])
Y= df['Class']
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3,stratify=Y)
y_train.value_counts()
```

```python
#building Isolation Forest model
#Isolation forest is an unsupervised algorithm that needs no labels to identify the anomalies in the data which in this case are the @fraud cases
# Isolation forest specially for Anomaly Detection
iso_forest = IsolationForest(n_estimators=100, max_samples=256,contamination=0.2, random_state=20)
iso_forest.fit(x_train)
score_anomalies = iso_forest.decision_function(x_train)
score_anomalies
```

```python
# plotting graph to identify anomalie score.
plt.figure(figsize=(12, 8))
plt.hist(score_anomalies, bins=100)
```

```python
# threshold for the anomalies(-0.07) was tuned till the best roc_auc_score was got, then used on the test set to ensure model not overfitting.
# 
from sklearn.metrics import average_precision_score
average_precision_score(y_train==0,(-0.07 < score_anomalies))
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_train==0,(-0.07 < score_anomalies))
```

```python
f1_score(y_train==0,(-0.07 < score_anomalies))
```

Model accurately predicts normal cases as normal and anomalies(fraud cases) as anomalies.

```python
score_anomalies_test = iso_forest.decision_function(x_test)
```

```python
from sklearn.metrics import average_precision_score
average_precision_score(y_test==0,(-0.07 < score_anomalies_test))
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test==0,(-0.07 < score_anomalies_test))
```

```python
f1_score(y_test==0,(-0.07 < score_anomalies_test))
```

from the ROC_AUC score on the training and test  set, it can be seen that the model is PRECISE in detecting fraud cases and can therefore help in reducing transactional cost from fraud.
